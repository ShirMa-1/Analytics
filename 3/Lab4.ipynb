```python
import numpy as np
from scipy.stats import t
from scipy.stats import norm
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import io
import requests
import seaborn as sns
import statistics
from scipy.stats import pearsonr
from sklearn.metrics import mean_squared_error, mean_absolute_error
import random
%matplotlib inline
```

# Пример № 1

```python
df1 = pd.read_csv('e1.csv')
df1.head()
```

```python
fonds = df1['fonds']
product = df1['product'] 

plt.scatter(fonds, product)
plt.xlabel('Стоимость ОФ')
plt.ylabel('Средняя выработка на 1 работника')
```

```python
fonds = np.array(fonds).reshape(-1,1)
product = np.array(product).reshape(-1,1)

model = LinearRegression()

X_train, X_test, y_train, y_test = train_test_split(
    fonds,
    product,
    test_size = 0.2
)

model.fit(X_train, y_train)
predicted_y = model.predict(X_test)
print(mean_squared_error(predicted_y, y_test))
print(mean_absolute_error(predicted_y, y_test))
```

```python
plt.xlabel('Стоимость ОФ')
plt.ylabel('Средняя выработка на 1 работника')
plt.title('Линейная регрессия от одной переменной')

x = np.linspace(-50, 200, 100)
y = model.coef_*x + model.intercept_
y = np.transpose(y)

plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(0,25)
plt.ylim(0,70)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

Зависимость линейная: чем выше стоимость, тем больше средняя выработка на 1 работника
